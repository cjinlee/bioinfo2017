```python
seq_list = dict()
f_fa = open('insulin_pep.uniprot.fa','r')
for line in f_fa:
    if line.startswith('>'):
        seq_h = line.strip()
        seq_list[seq_h] = ''
    else:
        seq_list[seq_h] += line.strip()
f_fa.close()


human = seq_list['>sp|P01308|INS_HUMAN Insulin OS=Homo sapiens GN=INS PE=1 SV=1']
pig = seq_list['>sp|P01315|INS_PIG Insulin OS=Sus scrofa GN=INS PE=1 SV=2']
cow = seq_list['>sp|P01317|INS_BOVIN Insulin OS=Bos taurus GN=INS PE=1 SV=2']
mouse = seq_list['>sp|P01325|INS1_MOUSE Insulin-1 OS=Mus musculus GN=Ins1 PE=1 SV=1']
chick = seq_list['>sp|P67970|INS_CHICK Insulin OS=Gallus gallus GN=INS PE=1 SV=1']
xenla = seq_list['>sp|P12706|INS1_XENLA Insulin-1 OS=Xenopus laevis GN=ins-a PE=1 SV=2']
danre = seq_list['>sp|O73727|INS_DANRE Insulin OS=Danio rerio GN=ins PE=3 SV=1']


def matcher(meas_seq, std_seq):    
    import numpy as np
    score = np.zeros([len(meas_seq)+1,len(std_seq)+1])
    for i in range (0, len(meas_seq)+1):
        score[i,0] = -6*i
    for j in range (0, len(std_seq)+1):
        score[0,j] = -6*j
        
    for i in range (1, len(meas_seq)+1):
        for j in range (1, len(std_seq)+1):
            if meas_seq[i-1] == std_seq[j-1]:
                value = score[i-1, j-1]+5
            else:
                value = score[i-1, j-1]-2
            score[i,j] = max(score[i-1,j]-6, score[i,j-1]-6, value)
    
    if meas_seq == human:
        name_meas = "human"
    elif meas_seq == pig:
        name_meas = "pig"
    elif meas_seq == cow:
        name_meas = "cow"
    elif meas_seq == mouse:
        name_meas = "mouse"
    elif meas_seq == chick:
        name_meas = "chicken"
    elif meas_seq == xenla:
        name_meas = "Xenopus laevis"
    elif meas_seq == danre:
        name_meas = "Danio rerio"

    if std_seq == human:
        name_std = "human"
    elif std_seq == pig:
        name_std = "pig"
    elif std_seq == cow:
        name_std = "cow"
    elif std_seq == mouse:
        name_std = "mouse"
    elif std_seq == chick:
        name_std = "chichken"
    elif std_seq == xenla:
        name_std = "Xenopus laevis"
    elif std_seq == danre:
        name_std = "Danio rerio"
        
    print (name_meas)
    print (meas_seq)
    print (score)
    print ("The score of ", name_meas, " compared to ", name_std, " is ", score[len(meas_seq), len(std_seq)], "\n")
    

matcher(pig, human)
matcher(cow, human)
matcher(mouse, human)
matcher(chick, human)
matcher(xenla, human)
matcher(danre, human)
```
