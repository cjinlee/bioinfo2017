```python
# get the information
seq_list = dict()
f_fa = open('insulin_pep.uniprot.fa','r')
for line in f_fa:
    if line.startswith('>'):
        seq_h = line.strip().lstrip('>').split()[0]
        seq_list[seq_h] = ''
    else:
        seq_list[seq_h] += line.strip()
f_fa.close()

human = seq_list['sp|P01308|INS_HUMAN']
pig = seq_list['sp|P01315|INS_PIG']
bovin = seq_list['sp|P01317|INS_BOVIN']
mouse = seq_list['sp|P01325|INS1_MOUSE']
chick = seq_list['sp|P67970|INS_CHICK']
xenla = seq_list['sp|P12706|INS1_XENLA']
danre = seq_list['sp|O73727|INS_DANRE']

# compute the score
def compute(subject, query, subject_start, query_start, length):
    score = 0
    for i in range(0,length):
        subject_base = subject[subject_start + i]
        query_base = query[query_start + i]
        if subject_base == query_base:
            score = score + 5
        else:
            score = score - 2
    return score

# for print
def print_match(subject, query, subject_start, query_start, length):
    print(str(subject_start) + (' ' * length) + str(subject_start+length))
    print(' ' + subject[subject_start:subject_start+length])
    print(' ' + query[query_start:query_start+length])
    print(str(query_start) + (' ' * length) + str(query_start+length))
    print("-" * 75)

# Try and Find the best match
def try_find(subject, query, score_limit):
    for subject_start in range(0,len(subject)):
        for query_start in range(0,len(query)):
            for length in range(0,len(query)):
                if (subject_start + length < len(subject) and query_start + length < len(query)):
                    score = compute(subject, query, subject_start, query_start, length)
                    if (score >= score_limit):
                        print("Score : " + str(score))
                        print_match(subject, query, subject_start, query_start, length)

print("Human and pig")
try_find(human, pig, 295)
print("Human and bovin")
try_find(human, bovin, 294)
print("Human and mouse")
try_find(human, mouse, 253)
print("Human and chick")
try_find(human, chick, 190)
print("Human and xenla")
try_find(human, xenla, 131)
print("Human and danre")
try_find(human, danre, 102)
```
