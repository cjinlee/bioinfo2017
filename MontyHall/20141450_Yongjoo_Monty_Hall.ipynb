```python
import random

game_n = 100
remain_win = 0
change_win = 0
for i in range(0,game_n):
    door_list = ['A','B','C']
    goat_list = ['A','B','C']
    car = random.choice(door_list)
    goat_list.remove(car)
    i_pick = random.choice(door_list)
    
    ## code from 'http://stackoverflow.com/questions/9915339/how-can-i-ignore-valueerror-when-i-try-to-remove-an-element-from-a-list'
    try:
        goat_list.remove(i_pick)
    except ValueError:
        pass  # do nothing!
    ##
    
    monty_open = random.choice(goat_list)
    door_list.remove(monty_open)
    
    #remain
    i_remain = i_pick
    if car == i_remain:
        remain_win += 1
        
    #change
    door_list.remove(i_pick)
    i_change = door_list[0]
    if car == i_change:
        change_win += 1
        
    print(car, monty_open, i_remain, i_change, remain_win, change_win)
    
remain_rate = remain_win / game_n
change_rate = change_win / game_n
print(remain_rate, change_rate)
```
